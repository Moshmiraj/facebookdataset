```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Data Collection
data = {
    'Programming Language': ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'Ruby', 'Go', 'Swift'],
    'Annual Salary (USD)': [120000, 115000, 110000, 105000, 110000, 100000, 115000, 125000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 2: Calculate Weekly Sales
# Calculate weekly sales (assuming 52 weeks in a year)
df['Weekly Sales (USD)'] = df['Annual Salary (USD)'] / 52

# Step 3: Data Exploration
print("First few rows of the dataset with weekly sales:")
print(df.head())  # Display the first few rows

print("\nBasic statistics of weekly sales:")
print(df['Weekly Sales (USD)'].describe())  # Basic statistics for weekly sales

# Step 4: Data Visualization using Matplotlib
plt.figure(figsize=(10, 6))

# Create a bar plot for weekly sales
plt.barh(df['Programming Language'], df['Weekly Sales (USD)'], color='skyblue')
plt.title('Weekly Sales by Programming Language')
plt.xlabel('Weekly Sales (USD)')
plt.ylabel('Programming Language')
plt.grid(axis='x')  # Add grid lines for better readability
plt.show()  # Display the plot
```

```python
import pandas as pd

# Sample data
data = {
    'Programming Language': ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'Ruby', 'Go', 'Swift'],
    'Annual Salary (USD)': [120000, 115000, 110000, 105000, 110000, 100000, 115000, 125000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Finding: Salary Insights
highest_salary = df[df['Annual Salary (USD)'] == df['Annual Salary (USD)'].max()]
print("Finding: Python has the highest salary among programming languages.")
print(f"Highest Salary: {highest_salary.to_string(index=False)}")

# Implication: Demand for Python
print("\nImplication: The high salary for Python suggests a growing demand in industries such as data science, machine learning, and backend development.")
```
